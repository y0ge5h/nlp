# Logistic Regression from Scratch - Gradient Descent and MLE

```python
import pandas as pd
import numpy as np
```

```python
def sigmoid(X, theta):
    z = np.dot(X, theta)
    sigmoid_value = 1 / (1 + np.exp(-z))
    return sigmoid_value
```

```python
def gradient_descent(X, h, y):
    gradient = np.dot(X.T, (h - y)) / y.shape[0]
    return gradient
def update_weight_loss(theta, learning_rate, gradient):
    new_theta = theta - learning_rate * gradient
    return new_theta
```

```python
data = pd.read_csv('sample_data.csv')
X = data.drop('Y', axis=1)
Y = data['Y']
```

```python
data.columns.values
```

```python
intercept = np.ones((X.shape[0], 1)) 
X = np.concatenate((intercept, X), axis=1)
print (X.shape)
theta = np.zeros(X.shape[1])
```

```python
iterations = 100
```

```python
for i in range(iterations):
    h = sigmoid(X,theta)
    gradient = gradient_descent(X, h, Y)
    print('old theta: ', theta)
    theta = update_weight_loss(theta, 0.1, gradient)
    print ('new theta: ', theta)
    print ('*************************')
```

## Model Evaluation

```python
predictions = sigmoid(X, theta)
```

```python
predictions
```

```python
f = pd.DataFrame(np.around(predictions, decimals=6)).join(Y)
f['pred'] = f[0].apply(lambda x : 0 if x < 0.5 else 1)
print("Accuracy (Loss minimization):")
f.loc[f['pred']==f['Y']].shape[0] / f.shape[0] * 100
```
